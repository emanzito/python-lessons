# **Example with possible data base**

```python
above_18 = True #from data base
has_discount = False #from data base

if above_18 or has_discount: #Just needs one to be true or false
    print("Ticket Price: 12€")
else:
    if above_18 and has_discount: #Needs both to be true or false
        print("Ticket Price: 8€")


#But this way the code doesn't have logic because else only works if first condition is false,
# which is not the case, so the best code is: 

if above_18 and has_discount:
    print ("Ticket Price: 8€")
else: 
    print ("Ticket Price: 12€")

```

**# Arithmetic**

```python
a = 10
b = 5

sum = a + b 
subtraction = a - b
multiplication = a * b  
division = a / b #gives an exact number 
floor_division = a // b #gives an integer number
modulus = a % b #gives the rest of a division
exponentiation = a ** b #b^a: 5^10


print (sum)
print (subtraction)
print (multiplication)
print (division)
print (floor_division)
print (modulus)
print (exponentiation)
```
