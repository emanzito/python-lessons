## **Arithmetic**

```python
a = 10
b = 5

sum = a + b
subtraction = a - b
multiplication = a * b
division = a / b #gives an exact number
floor_division = a // b #gives an integer number
modulus = a % b #gives the rest of a division
exponentiation = a ** b #b^a: 5^10


print (sum)
print (subtraction)
print (multiplication)
print (division)
print (floor_division)
print (modulus)
print (exponentiation)
```

## **Comparison**

```python
a = 10
b = 5

equal = a == b #Returns True if both values are equal
not_equal = a != b #Returns True if both values are different
greater_than = a > b #Returns True if a is bigger or greater than b
less_than = a < b #Returns True if a is smaller or less than b
greater_equal = a >= b #Returns True if a is greater than or equal to b 
less_or_equal = a <= b #Returns True if a is less than or equal to b

print (equal)
print (not_equal)
print (greater_than)
print (less_than)
print (greater_equal)
print (less_or_equal)
```

## **Logical**

```python
a = 10
b = 5

result_and = (a > 5) and (b < 10) #Both conditions need to be true
result_or = (a > 5) or (b > 10) #Just one condition needs to be true
result_not = not(a > 5) #Inverts the condition, if it's true becomes false and vice versa

print (result_and)
print (result_or)
print (result_not)
```

## **Exercise - For**

stock = [12, 5, 20, 3, 8, 15]

Objective:
    Use for to run the list   
    inside the circle, verify if the quantity is below 10   
    if it is, print - Alert, low stock (only [quantity] units!)   
    if it is 10 or more, print - Stock ok: [quantity] units!

```python
stock = [12, 5, 20, 3, 8, 15]

for stock_quantity in stock:
    if stock_quantity < 10:
        print (f"Alert, low stock (only  {stock_quantity}  units left)!")
    else:
        print (f"Stock ok: {stock_quantity} units available.")
```

## **Exercise - While**

Make a warning to a user of the cellphone when the battery hits below 5%      

Use the while, while the battery > 0          
Inside the cycle, print "Battery: {battery}%"          
Remove 1 to the battery each cycle           
In the end print: "Phone turned off           

```python
battery = 5

while battery > 0:
    print (f"Battery level: {battery} ")
    battery -= 1
    if battery == 0:
        print ("Battery empty. Please recharge.")
```

## **Another exercise - While**

Create a numeric safe

Objective:
    Create a variable called "correct_pin" with the value "1234"            
    Create a variable "attempt" that starts empty: " "          
    Use the cycle while that continues to run while "attempt" is different from the "correct_pin"         
                    
Inside the cycle:             
    Ask the user to type the PIN and save the try in "attempt"            
    If the PIN is wrong, print: ("Access denied. Try again")                 
               
Outside the cycle:              
    If the PIN is correct: ("Access granted, Welcome!)               

```python
correct_pin = "1234"
attempt = ""

while attempt != correct_pin:
    attempt = input("Enter the PIN to the safe:")
    if attempt == correct_pin:
        print("Access granted, Welcome!")
    else: 
        print("Access denied. Try again.")
```

## **Exercise - For && While**

Shop Cart System

We don't know how many products the client wants to buy

After the client finishes, we want to list everything that he bought and apply a discount. 


Part 1:
    Create an empty list called: "cart"           
    Print a list with all the products the client can buy and each price.                
    Create a cycle which asks the name of the product          
    If the user writes "End" the cycle stops             
    If not, keep asking to add products              

Part 2        
    With the list full, print each product of the list in uppercase.          
    Print how much products the client bought

```python
products = ["Screen 4k", "Gaming Mouse", "Mechanical keyboard", "Headphones", "Office Chair", "Webcam", "HDMI Cable 2m", "SSD 1TB", "Portable Support" ]
prices = [350, 45.90, 89, 59.99, 120, 15.50, 65, 12.90, 55, 25]

#Print the available products and the prices of each one
for i in range(len(products)):
    products_list = products[i]
    products_prices = prices[i]
    print(f"{products_list}: {products_prices}€")

print("-" * 30)

#Choosing which product the client will add to the cart
user_cart = []
gross_total = 0
cart = ""

while cart != "End":
    cart = input("Which product you wanna buy? Type End to finish your cart!")
    if cart in products:
        index = products.index(cart)
        product_price = prices[index]

        user_cart.append(cart)
        gross_total += product_price
        print(f"Added {cart} ({product_price}€) to your cart")
    else:
        print("Product not found. Please check the name (Case sensitive).")        

#Calculate discount based on the items quantity
numb_items = len(user_cart)
if numb_items > 0:
    discount_percent = numb_items
    discount_value = gross_total * (discount_percent / 100)
    total_final = gross_total - discount_value

    print("\n" + "="*30) 
    print(f"Items bought: {numb_items}")
    print(f"Gross total: {gross_total}")
    print(f"Discount: ({discount_percent}% -{discount_value:.2f}€)")
    print(f"Final price: {total_final:.2f}€")
    print("="*30)
else:
    print("Empty cart. See you next time!")




```
