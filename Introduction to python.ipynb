## **Arithmetic**

```python
a = 10
b = 5

sum = a + b
subtraction = a - b
multiplication = a * b
division = a / b #gives an exact number
floor_division = a // b #gives an integer number
modulus = a % b #gives the rest of a division
exponentiation = a ** b #b^a: 5^10


print (sum)
print (subtraction)
print (multiplication)
print (division)
print (floor_division)
print (modulus)
print (exponentiation)
```

## **Comparison**

```python
a = 10
b = 5

equal = a == b #Returns True if both values are equal
not_equal = a != b #Returns True if both values are different
greater_than = a > b #Returns True if a is bigger or greater than b
less_than = a < b #Returns True if a is smaller or less than b
greater_equal = a >= b #Returns True if a is greater than or equal to b 
less_or_equal = a <= b #Returns True if a is less than or equal to b

print (equal)
print (not_equal)
print (greater_than)
print (less_than)
print (greater_equal)
print (less_or_equal)
```

## **Logical**

```python
a = 10
b = 5

result_and = (a > 5) and (b < 10) #Both conditions need to be true
result_or = (a > 5) or (b > 10) #Just one condition needs to be true
result_not = not(a > 5) #Inverts the condition, if it's true becomes false and vice versa

print (result_and)
print (result_or)
print (result_not)
```

## **Exercise - Conditional Structures**

age = 18

Rules:     
    Under 12: Free          
    Between 12 & 17: 6€         
    +18:          
               student or coupon: 8€           
    not student and no coupon: 12€

```python
age = 18
is_student = False
has_coupon = False

if age < 12: 
    print ("Ticket Price: 0€")
elif age >= 12 and age <= 17:
    print ("Ticket Price: 6€")
elif age >= 18 and is_student or has_coupon:
    print ("Ticket Price: 8€")
else: 
    print ("Ticket Price: 15€")
    
```

## **Exercise - For**

ages = [8, 15, 22, 30, 11, 40]

Rules:      
    Under 12: 0€      
    Between 12 & 17: 6€    
    +18: 12€

Use for to:      
    - Calculate the price to each client     
    - The output to be like this:     
     Client age: 8 → Ticket Price: 0€

```python
ages = [8, 15, 22, 30, 11, 40]

for age in ages: 
    if age < 12:
        price = 0
    elif age >= 12 and age <= 17:
        price = 6
    elif age >= 18:
        price = 12

print(f"Age: {age}, Ticket Price: {price}€")
```
